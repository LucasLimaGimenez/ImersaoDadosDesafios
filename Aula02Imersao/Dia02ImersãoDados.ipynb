#Desafio 01 - Ordenar o gráfico countplot

```python
maiores_compostos = dados.query("composto in @cod_compostos")
display(maiores_compostos)
```

```python
contagem = maiores_compostos["composto"].value_counts().to_frame() #mostra os compostos que foram usados mais vezes e em formato de quadro para facilitar a visualização
display(contagem)
```

```python
plt.figure(figsize=(10,8)) #aumenta ou diminui o tamanho do gráfico a ser mostrado
ax = sns.countplot(x='composto',data = maiores_compostos,order=contagem.index) #tabela que recebe apenas um valor como busca estatistica nesse caso o x=composto ordenando do maior para o menor
ax.set_title("Top 5 compostos mais utilizados",fontsize = 18) #gera o titulo do gráfico
plt.show()
```

#Desafio 02 - Melhorar a visualização alterando tamanho da fonte ..

```python
plt.figure(figsize=(10,8)) 
ax = sns.countplot(x='composto',data = maiores_compostos,order=contagem.index,palette="rocket") #adiciona um tema ao gráfico
ax.set_title("Top 5 compostos mais utilizados",fontsize = 18) 
ax.set_xlabel("Quantidade usada",fontsize =  16) #gera o nome do eixo x do gráfico 
ax.set_ylabel("Nome do composto",fontsize = 16) #gera o nome do eixo y do gráfico
plt.show() #exibe o gráfico gerado
```

#Desafio 03 - Plotar as histogramas com seaborn


```python
sns.set()
plt.figure(figsize=(10,8))

ax = sns.histplot(x= dados['g-0'],data = dados.query("composto in @cod_compostos"),bins= 50) #gera um gráfico no modelo histograma utilizando o seaborn
ax.set_title("Gráfico em histograma",fontsize=20)
ax.set_xlabel("Quantidade de testes",fontsize=16)
ax.set_ylabel("Compostos utilizados",fontsize=16)

plt.show()
```

#Desafio 04 - Estudar sobre as estatisticas retornadas no .describe()

1- count: Contador do total do conjunto


2- mean: O valor médio;



3- std: Desvio padrão, é o valor da dispersão dos dados, ou seja, quanto um  valor aleatório tende a se desviar da média. Quanto menor, mais próximo o valor será da média;


4- min: O menor valor encontrado;


5- 25%: O primeiro quartil, o numero que divide o primeiro quarto dos numeros restante;


6- 50%: A mediana, o numero que divide o total de numeros ao meio;


7- 75%: O terceiro quartil, o numero que divide o ultimo quarto dos numeros restante;


8- max: O maior valor encontrado
