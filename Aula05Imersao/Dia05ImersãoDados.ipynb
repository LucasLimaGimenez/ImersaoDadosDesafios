#Utilizando dummies

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64') 
y = dados_combinados['ativo_moa'] 
x_treino , x_teste,y_treino , y_teste =train_test_split(x,y,test_size = 0.2,stratify=y,random_state = 376)

teste=[]
treino=[]

for i in range(1,15): #estrutura de repetição 
  modelo_arvore = DecisionTreeClassifier(max_depth = i) #quantos niveis vai ter a arvore
  modelo_arvore.fit(x_treino,y_treino)
  teste.append(modelo_arvore.score(x_teste,y_teste))  #calcula o quanto está acertando o teste
  treino.append(modelo_arvore.score(x_treino,y_treino)) #calcula o quanto está acertando o treino
```
