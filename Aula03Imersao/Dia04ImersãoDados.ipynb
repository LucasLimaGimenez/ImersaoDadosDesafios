#Realizando operações com n-moa

```python
dados_resultados['n_moa'] = dados_resultados.drop('id',axis=1).sum(axis = 1) #mostra quantos mecanismos de ação cada linha ativou/executou e guarda em uma nova coluna
dados_resultados['n_moa'].value_counts() #mostra os tipos existentes e a soma dos valores de cada tipo
```

```python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0 ) #cria uma coluna no final da tabela que dirá se possui ou não moas em cada linha
dados_resultados.head()
```

#Utilizando o merge



```python
dados_combinados = pd.merge(dados,dados_resultados[['id','n_moa','ativo_moa']],on='id') #realiza a combinação de duas tabelas com  a chave primária id
dados_combinados.head()
```

#Utilizando o query

```python
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts() #mostra que nenhum dos que possuem valor ativo_moa = true existem no tratamento com_controle logo todo tratamento com controle possui ativo_moa = 0(False) 
```

```python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts() #já nessa análise é possivel perceber que é muito mais frequente o ativo_moa receber valores quando o tratamento é com drogas
```

#Desafio 01 - Encontrar o top 10, das ações do MOA (inibidor,agonista...)

```python
#reset_index adiciona uma coluna no começa para que sejam mostrados de forma enumerada ao invés do index anterior
cont_moa = dados_resultados.drop(['id', 'n_moa', 'ativo_moa'], axis = 1).sum().sort_values(ascending = False).reset_index()
cont_moa.columns = ['Moa', 'contagem']
cont_moa
```

```python
acao = cont_moa['Moa'].str.split('_')
lista_acao=[]
for i in acao:
  lista_acao.append(i[-1])
lista_acao
```

```python
cont_moa['Ação'] = lista_acao
cont_moa
```

```python
top10 = cont_moa.groupby('Ação').agg({'contagem': 'sum'}).sort_values('contagem', ascending=False)
top10.head(10)
```
