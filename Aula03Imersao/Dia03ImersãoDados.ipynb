#Desafio 05 - Calcular e analisar a correlação entre os g e c.Refletir sobre os efeitos biológicos


```python
g = dados[dados['tratamento'] == 'com_droga'].loc[:,'g-0':'g-50']
c = dados[dados['tratamento'] == 'com_droga'].loc[:,'c-0':'c-50']

tb = pd.concat([g,c], axis = 1)
tb
```

```python
gc = tb.corr()
display(gc)
```

```python
filtrado = gc.loc[:'g-50','c-0':]
filtrado.shape
```

```python
f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(filtrado, cmap=cmap, center=0, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()
```

# *Reflefir sobre os efeitos biológicos:*
Podemos perceber que nos pontos mais vermelhos temos uma maior correlação. Vemos então uma alta relação/viabilidade do gene com a célula, ambos os valores vão para a mesma direção.

Já nos pontos mais azuis temos uma baixa correlação. Isso faz com que os seus valores mudem em direções opostas. Porém apenas com esse gráfico não podemos dizer se temos diminuição da viabilidade do gene com a célula e da célula com o gene, esse fator só é possível de ser visto no "lmpot()".

O gráfico acima nos ajuda a ver os genes que podem ganhar uma atenção maior.

Também podemos notar que muitos genes não tem uma grande correlação com a célula, podemos pensar que esses genes não estão envolvidos diretamente com as vias metabólicas das células escolhidas e não irão reagir da melhor maneira com o composto, então podemos pensar em descartar esses genes nos próximos testes.



#Desafio 06 - Estudar o código que plota a matriz de correlação (heatmap)

Parâmetros da função:

a) vmin, vmax -> qual o valor minimo (vmin) e maximo (vmax) que será considerado ao plotar a régua (legenda do lado direito do gráfico) e a opacidade das cores que indicam a correlação entre as variáveis;

b) cmap -> cor(es) que serão utilizadas na figura;

c) center -> em qual valor as cores serão mais opacas, isto é, serão menos visíveis e a correlação será considerada mais nula;

d) robust -> se valores minimos e maximos não forem passados e robusto for igual a True, os valores minínimos e máximos será calculado automaticamente utilizando o conceito de robust quantiles. Nesse sentido, ele irá setar valores minínimos e máximo de acordo com a distribuição de cada dataframe de correlação específico;

e) annot -> se True ele insere o valor da correlação em cada quadrado que corresponde à correlação de duas variáveis específicas etc.

Além destes parâmetros indicados acima, a função possui elementos para que possamos mexer em aspectos estéticos como a grossura da linha, formato dos textos da imagem etc. Sempre bom lembrar que o matplotlib pode ser utilizado junto ao seaborn se houver necessidade de ajustes que não são facilmente encontrados diretamente no sns.
